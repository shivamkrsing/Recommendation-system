```python
import pandas as pd
movie=pd.read_csv("movies.csv")
rating=pd.read_csv("ratings.csv")
```

```python
movie.head()
```

```python
rating.head()
```

```python
df=pd.merge(rating,movie,on="movieId")
df.head()
```

```python
df.groupby("title")["rating"].mean().sort_values(ascending=False).head()
```

```python
df.groupby("title")["rating"].count().sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(df.groupby("title")["rating"].mean())
ratings.head()
```

```python
ratings["Num of rating"]=pd.DataFrame(df.groupby("title")["rating"].count())
ratings.head()
```

```python
moviemat=df.pivot_table(index="userId",columns="title",values="rating").fillna(0)
moviemat.head()
```

```python
ratings.sort_values("Num of rating",ascending=False).head()
```

```python
forest_rating=moviemat['Forrest Gump (1994)']
forest_rating.head()
```

```python
similar_to_forest=moviemat.corrwith(forest_rating)
```

```python
corr_forest=pd.DataFrame(similar_to_forest,columns=["correlation"])
corr_forest.dropna(inplace=True)
corr_forest.head()
```

```python
corr_forest.sort_values("correlation",ascending=False).head()
```

```python
corr_forest=corr_forest.join(ratings["Num of rating"])
corr_forest.head()
```

```python
corr_forest[corr_forest["Num of rating"]>100].sort_values("correlation",ascending=False).head()
```

```python
Jurassic_rating=moviemat['Jurassic Park (1993)']
similar_to_Jurassic=moviemat.corrwith(Jurassic_rating)
corr_Jurassic=pd.DataFrame(similar_to_Jurassic,columns=["correlation"])
corr_Jurassic.dropna(inplace=True)
corr_Jurassic=corr_Jurassic.join(ratings["Num of rating"])
corr_Jurassic[corr_Jurassic["Num of rating"]>100].sort_values("correlation",ascending=False).head()
```
