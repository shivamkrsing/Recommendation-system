```python
R=df2["vote_average"]
v=df2["vote_count"]
c=df2["vote_average"].mean()
m=df2['vote_count'].quantile(0.70)
```

```python
df2["weighted_vote"]=(R*v+c*m)/(v+m)
```

```python
df2.head()
```

```python
dp=df2[["weighted_vote",'popularity']]
dp
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
a=scaler.fit_transform(dp)
movie=pd.DataFrame(a,columns=["weighted_vote",'popularity'])
movie
```

```python
movie["score"]=(movie["weighted_vote"]*40 + movie["popularity"]*60)/100
movie
```

```python
movie.sort_values(['score'],ascending=False)
```
